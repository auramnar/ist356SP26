```python
data1 = [
    {
        "name": "Alice",
        "age": 30,
        "address": {
            "street": "123 Main St",
            "city": "New York"
        },
        "hobbies": ["reading", "hiking"]
    },
    {
        "name": "Bob",
        "age": 25,
        "address": {
            "street": "456 Elm St",
            "city": "Los Angeles"
        },
        "hobbies": ["gaming", "cycling"]
    }
]
```

```python
# Q11
# How many rows and columns are in the following data frame? 
# A. 2 rows, 5 columns 

pd.json_normalize(data1)
```

```python
# Q 12 - C.	4 rows, 5 columns
# How many rows and columns are in the following data frame? 

# record_path flattens the nested list of hobbies. Tells pandas to make each hobby its own row.
# meta specifies the other fields to include in each row.
# ['address', 'street'] and ['address', 'city'] tell Pandas to extract nested fields in the address dictionary.
# errors='ignore' means avoids errors if any of the entries don’t contain one of those fields.

df = pd.json_normalize(data1, 
                       record_path='hobbies', 
                       meta=['name', 'age', ['address', 'street'], ['address', 'city']], 
                       errors='ignore')
df
```

Best practice - data lineage
When combining datasets, its a really good idea to retain data lineage, or a record of where the data came from. This can be done by added a column to each dataframe before concatenating.

```python
# Question 25
# when you perform arithmetic between two columns (Series), 
# the operation is vectorized. 
# This means pandas automatically matches up the values row-by-row and 
# performs the division for every employee simultaneously
# Numerator: employee_df['Tasks Completed'] targets the work output.
# Denominator: employee_df['Hours Worked'] targets the time input.
# Result: A new column named 'Productivity' is added to the DataFrame 
# containing the calculated rates

employee_df = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Cara', 'Dan'],
    'Deparatment': ['Sales', 'Sales', 'HR', 'HR'],
    'Tasks Completed': [50, 45, 30, 25],
    'Hours Worked': [40, 35, 40, 32]
})


employee_df['Productivity'] = employee_df['Tasks Completed'] / employee_df['Hours Worked']
employee_df
```
